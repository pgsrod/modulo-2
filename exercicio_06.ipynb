```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt

try:
    df = pd.read_csv('03.BaseDPEvolucaoMensalCisp.csv', encoding='ISO-8859-1 ', sep = ';')
    print(df.columns)

except Exception as e:
    print(f'Erro: {e}')
    exit()
```

```python
# 1. Quantidade de veículos recuperados por municipio

print(f'--- Ranking de Recuperação de Veículos por Município ---')

df_recuperacao_veiculos = df[['munic', 'recuperacao_veiculos']]

df_recuperacao_veiculos = df_recuperacao_veiculos.groupby('munic').sum().reset_index().sort_values(by='recuperacao_veiculos', ascending=False)

df_recuperacao_veiculos

```

 Perguntas para responder 

1. Existe grande diferença entre os primeiros municípios do ranking?

    Resposta: Analisando os cinco primeiros municípios do ranking, observa-se uma grande diferença entre o primeiro colocado, Rio de Janeiro, e os demais. Já os outros quatro municípios apresentam valores relativamente próximos de recuperação de veículos. Esse resultado sugere uma forte concentração de recuperações na capital em comparação com os demais municípios analisados.


```python
print(f'--- Ranking de Roubo de Veículos por Município ---')

df_roubo_veiculos = df [['munic', 'roubo_veiculo']]

df_roubo_veiculos = df_roubo_veiculos.groupby('munic').sum().reset_index().sort_values(by='roubo_veiculo', ascending=False)

df_roubo_veiculos
```

 Perguntas para responder 
 
 2. Compare com o ranking de roubo de veículos. O município que mais tem roubos também lidera em recuperações?
 
Resposta: Ao comparar os rankings de recuperação e de roubos de veículos, observa-se que os cinco primeiros municípios são os mesmos em ambos os rankings: Rio de Janeiro, Duque de Caxias, São Gonçalo, São João de Meriti e Nova Iguaçu.

No entanto, há uma pequena diferença na posição de dois municípios. São João de Meriti aparece em 4º lugar no ranking de recuperação de veículos, mas ocupa o 5º lugar no ranking de roubos. Já Nova Iguaçu aparece em 5º lugar na recuperação, mas ocupa o 4º lugar no ranking de roubos.

Dessa forma, conclui-se que, embora os mesmos municípios concentrem tanto roubos quanto recuperações de veículos, há pequenas variações na posição ocupada por alguns deles nos dois rankings.

```python
print(f'--- Ranking de Recuperação de Veículos ---')

df_recuperacao_veiculos = df[['ano', 'recuperacao_veiculos']]

df_recuperacao_veiculos = df_recuperacao_veiculos.groupby('ano').sum().reset_index().sort_values(by='recuperacao_veiculos', ascending=False)

df_recuperacao_veiculos
```

Perguntas para responder

1. Qual foi o ano com maior número de recuperações de veículos?

Resposta: O ano com o maior número de recuperações de veículos foi 2017, com um total de 37.630 veículos recuperados.

```python

print(f'--- Ranking Anual de Recuperação de Veículos ---')

df_recuperacao_veiculos = df[['ano', 'recuperacao_veiculos']]

df_recuperacao_veiculos = (df_recuperacao_veiculos.groupby('ano').sum().reset_index().sort_values(by='ano', ascending=False))

df_recuperacao_veiculos
```

Perguntas para responder

2. Existe alguma tendência de crescimento ou queda a longo dos anos?

Resposta: Ao analisar o ranking anual de recuperação de veículos, observa-se uma tendência de queda ao longo dos anos. Desde o pico em 2017, o número de veículos recuperados vem diminuindo gradualmente até 2024, indicando redução nas recuperações no período analisado.

```python
print('--- Ranking Anual de Roubo e Recuperação de Veículos ---')

df_anual_veiculos = df[['ano', 'roubo_veiculo', 'recuperacao_veiculos']]

df_anual_veiculos = (df_anual_veiculos.groupby('ano').sum().reset_index().sort_values(by='ano', ascending=False))

df_anual_veiculos
```

Perguntas para Responder

3. esse comportamento acompanha o crescimento ou queda os roubos de veículos?

Respostas: Entre 2017 e 2024, tanto os roubos quanto as recuperações de veículos apresentam uma tendência geral de queda. Os roubos caíram de 54.366 para 15.327, enquanto as recuperações caíram de 37.630 para 9.626 veículos.

A queda das recuperações acompanha a redução dos roubos, embora a proporção de veículos recuperados em relação aos roubados tenha se mantido relativamente estável, com leve variação ao longo dos anos.
