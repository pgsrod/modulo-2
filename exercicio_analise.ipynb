```python

import pandas as pd 

import numpy as np

```

```python

df_vendas = pd.read_csv('01.amazon_sales_dataset.csv')
df_vendas.columns
```

```python
# definir a coluna que será analisada

coluna_alvo = 'discount'

# calculando as medidas centrais e posição

total_desconto = df_vendas[coluna_alvo].sum()
media_desconto = df_vendas[coluna_alvo].mean()
desconto_minimo = df_vendas[coluna_alvo].min()
desconto_maximo = df_vendas[coluna_alvo].max()

q1 = np.percentile(df_vendas[coluna_alvo], 25)

q2 = np.percentile(df_vendas[coluna_alvo], 50)

q3 = np.percentile(df_vendas[coluna_alvo], 75)

```

```python
delta_media_mediana = media_desconto - q2

distancia_percentual = media_desconto / q2

assimetria = df_vendas[coluna_alvo].skew()

```

```python

print(f'Para analisarmos os descontos realizados nas vendas do período, iremos verificar os resumos de descontos aplicados abaixo:')

print('\n--- Resumo de Descontos Realizados ---')

print(f'1. Valor Total de descontos: $ {total_desconto:,.2f}')

print(f'2. Desconto Médio aplicado: $ {media_desconto:,.2f}')

print(f'3. Maior desconto aplicado em vendas: $ {desconto_maximo:,.2f}')

print(f'4. Menor desconto possível em uma venda: $ {desconto_minimo:,.2f}')
```

```python
print(f'Abaixo o resumo de descontos aplicados, tais quais a o valor de média, mediana e o delta de descontos aplicados')


print('\n--- Relatorio de Descontos em Vendas Realizadas ---\n')

print(f'A Média de descontos é de $ {media_desconto:,.2f}')

print(f'A mediana (valor central) de descontos é de $ {q2} ')

print(f'O delta (Diferença) entre elas é de: $ {delta_media_mediana:,.2f}')

```

```python
print(f'Para analisarmos a tendência do gráfico de descontos, iremos verificar abaixo o relatório de análise comportamental e assimetria dos descontos aplicados no relatório de vendas')

print('\n --- Análise Comportamental ---\n')

print(f'Distancia relativa entre elas: {distancia_percentual:.2f}')

if abs(distancia_percentual-1)<0.10:
    print(f'INFERÊNCIA: Baixa Dispersão. A média é confiável para representar a coluna.')

elif abs(distancia_percentual-1)<0.25:
    print(f'INFERÊNCIA: Dispersão Moderada. Fique atento aos extremos, média começa a mentir.')

else:
    print(f'INFERÊNCIA: Alta dispersão. A média NÃO é confiável, foque na mediana')
    

print(f'Grau de assimetria (skewness); {assimetria}')

if assimetria > 0.5:
    print(f'INFERÊNCIA (Assimetria Positiva): A média de desconto é maior que a mediana. Significa que poucos descontos "agressivos" são dados')

elif assimetria < -0.5:
    print(f'INFERÊNCIA (Assimetria Negativa): A média de desconto é menor que a mediana. Significa que são dados descontos altos com frequência, mas alguns descontos muito pequenos baixam a média.')

else:
    print(f'INFERÊNCIA (Simétrica): Distribuição Equilibrada. Média e Mediana estão bem próximas. A estratégia de descontos da empresa está padronizada e previsível.')

print(f'Conclusão da Análise: A operação de descontos é uniforme, ou seja, os descontos aplicados não distorcem os números')
```
