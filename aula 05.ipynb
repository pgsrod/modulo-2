```python
import pandas as pd 

import numpy as np


```

```python
df_vendas = pd.read_csv('01.amazon_sales_dataset.csv')
df_vendas.columns
```

```python
df_vendas.dtypes
```

```python
df_vendas.info()
```

```python
coluna_alvo = 'total_sales'

# medidas de tendência centrais e posição

total_vendas = df_vendas[coluna_alvo].sum()
media_valor_vendas = df_vendas[coluna_alvo].mean()
venda_minima = df_vendas[coluna_alvo].min()
venda_maxima = df_vendas[coluna_alvo].max()
```

```python
# medidas de posição 

q1 = np.percentile(df_vendas[coluna_alvo], 25)

# q2 = mediana
q2 = np.percentile(df_vendas[coluna_alvo], 50)

q3 = np.percentile(df_vendas[coluna_alvo], 75)
```

```python
delta_media_mediana = media_valor_vendas - q2

distancia_percentual = media_valor_vendas / q2


# Caso 1. Distancia entre 
# Caso 2. Distancia entre 10% e 25 % (0.75 ~ 0.9 / 1.10 ~ 1.25)
# Caso 3. Distancia > 25% (<0.75 / > 1.25)
# R = Não confie na média, ela está distorcida pelos extremos 
```

```python
# medidas de dispersão 

# 1. Amplitude
amplitude = venda_minima - venda_minima

# 2. variancia e Desvio Padrão -  Medem a distancia média dos pontos até a Média
# Se for um valor alo -> Maior volatilidade
variancia = df_vendas[coluna_alvo].var()
desvio_padrao = df_vendas[coluna_alvo].std()

# 3. IQR -> 50% "internos" dos dados (25% ~ 75%) 
# Muito robusto a ouliers
iqr = q3 - q1

# 4. Coeficiente de Variação (CV)
# Muito usado para comparar volatilidades de 'grandezas' diferentes (colunas diferentes)
# OBS.: Multiplicado por 100 para ser descito como porcentagem 
cv = (desvio_padrao/media_valor_vendas)*100
```

```python
# assimetria (skewness)

assimetria = df_vendas[coluna_alvo].skew()

#  Caso 1. Assimetria entre -0.5 e 0.5
# R = dados equilibrados  
# Caso 2. Assimetria > 0.5 (positiva)
# R = valores altos sã ratos, porém são tão grandes que distorcem o calculo
# Caso 3. Assimetria < - 0.5 ( Negativa)
# R = Menos valores distorcem o calculo para baixo 

# kurtosis (curtose)
# mede o 'peso' das caudas -> frequencia e gravidade dos outliers
curtose = df_vendas[coluna_alvo].kurtosis()

```

```python
print('\n--- Resumo Executivo de Vendas ---')

print(f'1. Volume Total (Soma das Vendas): $ {total_vendas:,.2f}')
# Representa o tamanho total da nossa operação no período 

print(f'2. Gasto Médio: $ {media_valor_vendas:,.2f}')
#  valor esperado que um cliente gaste em média conosco

print(f'3. Maior Venda: $ {venda_maxima:,.2f}')
# Nosso recorde de venda. Investigar o perfil do cliente.

print(f'4. Menor Venda: $ {venda_minima:,.2f}')
# Menor valor vendido. Investigar se há anomalias ou fraude.
```

```python
print('--- Relatorio Estatistico de Negócio ---')

print(f'A Média de vendas é $ {media_valor_vendas:,.2f}')

print(f'A mediana (valor central) das vendas é $ {q2} ')

print(f'O delta (Diferença financeira) entre elas é: $ {delta_media_mediana:,.2f}')

print('\n --- Analise Comportamental ---\n')

print(f'Distancia relativa entre elas: {distancia_percentual:.2f}')

if abs(distancia_percentual-1)<0.10:
    print(f'INFERÊNCIA: Baixa Dispersão. A média é confiável para representar a coluna.')

elif abs(distancia_percentual-1)<0.25:
    print(f'INFERÊNCIA: Dispersão Moderada. Fique atento aos extremos, média começa a mentir.')

else:
    print(f'INFERÊNCIA: Alta dispersão. A média NÃO é confiável, foque na mediana')
    

print(f'Grau de assimetria (skewness); {assimetria}')

if assimetria > 0.5:
    print(f'INFERÊNCIA (Assimetria Positiva): Nossas vendas tem uma cauda longa para a direita. Os grandes clientes são minoria, todavia faturam valores expressivos puxando a venda média para cima')

elif assimetria < -0.5:
    print(f'INFERÊNCIA (Assimetria Negativa): Nossas vendas tem uma cauda longa para a esquerda. O valor médio está sendo puxado para baixo por um grande volume de vendas baratas, talvez por liquidação')

else:
    print(f'INFERÊNCIA (Simétrica): Distribuição Equilibrada. Média e Mediana estão bem próximas.')
```

```python
print(f'--- Dispersão ---')
print(f'Amplitutide Total: $ {amplitude}')
# Amplitude no negócio: Diferença entre a maior e menor venda. 
print(f'Desvio de Padrão: $ {desvio_padrao}')
# Desvio no negócio: Nível de dispersão dos valores de venda
print(f'Coeficiente de variação: {cv:.2f}%')
if cv > 30: 
    print('INFERÊNCIA: Volatilidade Alta. Não há consistência no faturamento diário')
else:
    print('INFERÊNCIA: Volatidade Baixa. Vendas são consistência e previsíveis.')

print(f'Intervalor Interquartil (IQR): $ {iqr:.2}')

print('INFERÊNCIA: Esta é a variação de gastos do nosso consumidor padrão, ignorando clientes que compram muito pouco ou valores muito altos')

# curtose -> curva normal (curtose = 3)
# curtose no pandas -> Já trata do excesso sobre a curva normal
# No Pandas .kurtosis() = 0 -> Mesocúrtica

print(f'Curtose (Excesso - Fisher): {curtose:.4f}')
if curtose > 0.5:
    print('INFERÊNCIA: Distribuição Leptocúrtica. Pico agudo, caudas grossas')
    print('Sobre o negócio: A maioria das vendas estão no centro, todavia temos uma altíssima probabilidade de eventos extremos (outliers)')

elif curtose < -0.5:
    print(f'INFERÊNCIA: Distribuição Platicúrtica. Pico suave (achatado), caudas finas')
    print('Sobre o negócio: As vendas estão muito bem distribuídas, chance baixa de ter eventos extremos (outliers) ')

else:
    print(f' INFERÊNCIA: Distribuição Mesocúrtica. Pico mediano, caudas normais')
    print('Sobre o negócio: Comportamento normal e totalmente previsível')
```

```python
# df_vendas é a base
# agrupar por qual coluna? category
# o que iremos calcular? total_sales

df_categoria_agrupada = df_vendas.groupby('category')['total_sales'].sum().reset_index()

df_categoria_ordenada = df_categoria_agrupada.sort_values('total_sales' , ascending=False)

print(df_categoria_ordenada)
```
